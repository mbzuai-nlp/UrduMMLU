# Urdu MMLU — Spot Check

Browse samples from `mcqs.json` (the clean canonical version).

Self-contained: `print_urdu()` is defined inline below — no external dependency.

## Setup

```python
import json, random, html
from pathlib import Path
from collections import Counter
from IPython.display import display, HTML

# One-time CSS injection: load Nastaliq from Google Fonts.
display(HTML('<link href="https://fonts.googleapis.com/css2?family=Noto+Nastaliq+Urdu:wght@400;700&display=swap" rel="stylesheet">'))

def print_urdu(text):
    """Render text as RTL Nastaliq in Jupyter output. No card, just text."""
    display(HTML(
        f'<div dir="rtl" style="'
        f'font-family:\'Noto Nastaliq Urdu\',\'Jameel Noori Nastaleeq\',serif;'
        f'font-size:18px;line-height:2;margin:2px 0;color:inherit">'
        f'{html.escape(str(text))}</div>'
    ))

data = json.loads((Path.cwd() / 'mcqs.json').read_text(encoding='utf-8'))
by_sub = {}
for it in data:
    by_sub.setdefault(it.get('subdomain', '?'), []).append(it)

def show(item):
    """Print one MCQ: tag line (plain), question + options (Nastaliq)."""
    print(f"id={item['id']}  sub={item.get('subdomain')}  level={item.get('level') or '-'}")
    print_urdu(item['question'])
    for k, v in (item.get('options') or {}).items():
        print_urdu(f'{k}. {v}')
    co = item.get('correct_option')
    if co:
        print_urdu(f'✓ {co}')
    print()

print(f'Loaded {len(data):,} MCQs · {len(by_sub)} subdomains')
```

## Random sample

```python
random.seed(1)
for it in random.sample(data, 5):
    show(it)
```

## Mathematics — exponents, log subscripts, LaTeX

```python
random.seed(2)
for it in random.sample(by_sub['mathematics'], 5):
    show(it)
```

## Physics — `m1` / `m2` etc. preserved (no false superscript)

```python
random.seed(3)
for it in random.sample(by_sub['physics'], 5):
    show(it)
```

## Chemistry — `H2O`, `CO2`, `NaCl` untouched

```python
random.seed(4)
for it in random.sample(by_sub['chemistry'], 5):
    show(it)
```

## Urdu literature

```python
random.seed(5)
for it in random.sample(by_sub.get('urdu literature', []), 5):
    show(it)
```

## Pakistan studies — answered set

```python
answered_pak = [it for it in by_sub.get('pakistan studies', []) if it.get('correct_option')]
random.seed(6)
for it in random.sample(answered_pak, 5):
    show(it)
```

## Quick stats

```python
answered = sum(1 for it in data if it.get('correct_option'))
print(f'Total MCQs:   {len(data):,}')
print(f'  answered:   {answered:,}')
print(f'  unanswered: {len(data) - answered:,}\n')
print('Top 10 subdomains:')
for sub, n in Counter(it.get('subdomain') for it in data).most_common(10):
    print(f'  {sub:<35} {n:>6}')
```
